```python
import numpy as np
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
X.shape
```

```python
y
```

```python
w = 0
lr = 0.01

for epoch in range(10):
    y_pred = w * X
    loss = np.mean((y - y_pred)**2)
    w = w - lr * np.mean(2 * X * (y_pred - y))
    print(f"epoch {epoch}, loss {loss:.2f}, w {w:.2f}")
```
